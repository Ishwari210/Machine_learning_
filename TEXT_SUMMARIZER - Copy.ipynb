```python
import pandas as pd

data = {
    'id': [1, 2, 3],
    'tweet': [
        "I love the new design! So sleek and fast!",
        "This is the worst update ever üò†",
        "It's okay, not bad but not great either."
    ]
}

df = pd.DataFrame(data)
```

```python
from textblob import TextBlob

def get_sentiment(text):
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity  # -1 to 1
    if polarity > 0:
        return 'Positive'
    elif polarity < 0:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment'] = df['tweet'].apply(get_sentiment)
print(df[['tweet', 'Sentiment']])
```
