```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,"g--",label="y=x")
plt.title("Simple Graph")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

**PLOT**

* x:Array to plot as per X axis
* y:Array to plot as per y axis
* color:Color of the line ["blue","red","g","#FFF","#000"]
* linestyle:Dashed Line(--),Line(--),Dotted Line(:)
* Marker:To mark it on the plot('0','x','*')
* linewidth:Width of the line(1.0,2.0)
* label:To add label to the axis
* alpha:For controlling capacity/transparency
* grid flag:To show the grid or not

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x,y,"g--",linewidth="4",marker="o",markersize="15",label="y=x")
plt.title("Simple Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

**SCATTER PLOT**
* X:Array for x axis
* Y:Array for y axis
* S:Scalar, To set the size of the points
* C:Color, To set the color of the markers
* Alpha:Opacity/transparency controlling, [0,0-1,0]
* label:To give label to the graphs
* cmap:To set the gradient

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.scatter(x,y,s=100,c="blue",alpha=1,cmap="viridis",label="scatter")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[1,2,3,4]
plt.scatter(x,y,s=100,c="blue",alpha=0.1,label="scatter")
plt.legend()
plt.show()
```

**BAR CHART**
* X:Array which represent bars along x axis
* Height:Height of each bar
* Width:The width of the bar
* Color:To set color of the bar
* align:To align the bar("center","edge")
* label:To assign label to the bar

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
barheight=[10,20,30,40]
plt.bar(x,barheight,color="green",label="BarChart")
plt.legend()
plt.show()
```

**Pie Chart**

* X:This is an array for each portion
* Labels:This is an array for labeling each portion
* Color:This is an array for coloring each portion
* autopct: Formatting the labels
* startangle: For determining angle of first slice
* shadow: To add shadow effect to the piechart

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
pname=['A','B','C','D']
pcolors=['red','blue','green','yellow']
plt.pie(x, labels=pname, colors=pcolors, shadow=True)
plt.legend()
plt.show()
```

**HISTOGRAM**
* X:It's an array for plotting historical bars
* bins:Number of edges
* range:It is to limit the bins
* color:To set the color of bars
* alpha:Opacity/Transparency 0 to 1
* label:To label the chart
* Density:It will normalize the data

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,1,1,3,4,5]
plt.hist(x, bins=5, color="red", alpha=0.5, label="Histogram Data")
plt.legend()
plt.show()
```
