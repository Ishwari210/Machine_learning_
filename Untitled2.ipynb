```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
df=pd.read_csv(r'/content/admission_data.csv')
df.head()

```

```python
print(df.columns)
```

```python
df.loc[df['Chance of Admit '] >= 0.80,'Chance of Admit '] = 1
df.loc[df['Chance of Admit '] < 0.80,'Chance of Admit '] = 0
df['Chance of Admit ']
```

```python
df.loc[df['Chance of Admit '] >=0.80,'Chance of Admit']=1
df.loc[df['Chance of Admit'] < 0.80,'Chance of Admit']=0
df['Chance of Admit']
```

```python
x=df[['Chance of Admit','CGPA','GRE Score']]
y=df['TOEFL Score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print("x_train:",x_train.head())
print("x_test:",x_test.head())
print("y_train:",y_train.head())
print("y_test:",y_test.head())
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,30))
plot_tree(model,filled=True,rounded=True,feature_names=['Chance of Admit','CGPA','GRE Score'])
plt.show()
```
