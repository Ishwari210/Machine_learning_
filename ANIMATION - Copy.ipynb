**ANIMATION**
* FuncAnimation
* Fig: Figure which we will use to plot data set
* dpi: Dots per inch.Default(100.0)
* linewidth: width of edge of the frame
* tight_layout: To prevent plots from overlapping
* edgecolor: To set the color of edge for the frame
* frames: No. of frames which will be required for animation
* interval: milliseconds,per frame milliseconds
* blit: only the changed portion of the plot are redrawn.To improve the performance
* func: Function which will perform operations on dataset, which will again be plotted on figure
* repeat: To repeat the animation infinetly.(Boolean: True/False, Default:True)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

fig,ax=plt.subplots()
dataset=np.linspace(0,2*np.pi,100)
line,=ax.plot(dataset,np.sin(dataset))
def draw_line(frame):
  line.set_ydata(np.sin(dataset*frame)/100)
  return line
  ani=FuncAnimation(fig,draw_line,frames=100,interval=30)
  ani.save("Sin Wave.gif",fps=30,writer="pillow")
```

**SUBPLOTS**
* figsize: (h,w)

```python
from matplotlib.animation import ArtistAnimation
fig, ax=plt.subplots(figsize=(10,10))
countries=["India","USA","UK","Egypt","AUS"]
years=[2001,2002,2003,2004,2005,2006]
gdp=np.array([2,3,5.5,3.4,4])
rng=np.random.default_rng(10)
growth=rng.uniform(0.1,0.7,size=(len(years),len(countries)))
args=[]
def draw_bar():
  temp=[]
  for index,year in enumerate(years):
    gdp_1=(1+growth[index])
    bar_val=ax.barh(countries,gdp_1,color=['tab:red','tab:blue','tab:purple','tab:green','tab:cyan'])
    year_text=ax.text(0.95,0.1,f"Year:{year}",transform=ax.transAxes, fontsize=14, ha="right", va="bottom", color="black",weight="bold")
    print(year_text)
    temp.append([*bar_val,year_text])
  return temp
args=draw_bar()
ani=ArtistAnimation(fig,args,interval=100)
plt.show()


```
