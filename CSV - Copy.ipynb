**CSV and TEXT**

* Filepath: To select the file path from the directory, It could be string, pathlib, path, localpath, object, Buffer(StringIO)
* SEP: Separator/Delimiter, Sniffer Tool, it to detect the delimiter automatically
* delimiter:alternative to SEP
* delim_whitespace: It is used specify where '' is used as delimiter
* header: int, It is used to specify the row which is a header
* names: List, It is used to specify which column names to be used as columns
* index_col: It is used to select the column which will be used as a index
* dtype: List, It is used to specify the data types of each column
* data_parser: It is used to format the column which has the date as it's value
* true_value: List
* compression: gzip,zip,infer
* encoding:'ut8'
* iterator

**CONCATENATION**

```python
dataframe=pd.DataFrame({
    "longitude":[1,2,3,4],
    "latitude":[5,6,7,8],
    "housing_median_age":[12,23,44,33]

})
dataframe_2=pd.DataFrame({
    "longitude":[2,3,5,6],
    "latitude":[4,7,8,9],
    "housing_median_age":[11,22,33,44]
})
result=pd.concat([dataframe,dataframe_2])
print(result)
result=pd.concat([dataframe,dataframe_2],axis=1)
print(result)
```

**MERGE**

```python
dataframe=pd.DataFrame({
    "longitude":[1,2,3,4],
    "latitude":[5,6,7,8],
    "housing_median_age":[12,23,44,33]

})
dataframe_2=pd.DataFrame({
    "longitude":[2,3,5,6],
    "latitude":[4,7,8,9],
    "housing_median_age":[11,22,33,44]
})
inner_join=pd.merge(dataframe,dataframe_2,on="longitude",how="inner")
print(inner_join)
left_join=pd.merge(dataframe,dataframe_2,on="longitude",how="left")
print(left_join)
outer_join=pd.merge(dataframe,dataframe_2,on="longitude",how="outer")
print(outer_join)
```
