CONVERTING LIST OF SERIES TO DATAFRAME

```python
import numpy as np
import pandas as pd
series_1=pd.Series([1,2,3,4,5,6], index=['a','b','c','d','e','f'])
series_2=pd.Series([10,20,30,40,50,60], index=['a','b','c','d','e','f'])
data={
    "first":series_1,
    "second":series_2
}
data_frame=pd.DataFrame(data,index=['a','b'])
data_frame
```

CONVERTING DICT TO DATAFRAME

```python
dict_1=[1,2,3,4,5,6]
dict_2=[1.1,1.2,1.3,1.4,1.5,1.6]
dict={
    "first":dict_1,
    "second":dict_2
}
dataframe=pd.DataFrame(dict, index=['a','b','c','d','e','f'])
dataframe
```

CONVERTING RECORDS TO DATAFRAME

```python
data=[(1,2,3,4),(10,20,30,40)]
dataframe=pd.DataFrame(data)
dataframe
```

CONVERTING TUPLES TO DATAFRAME

```python
tuple={
    ("first_name","last_name"):{("Name","FullName"):"Ashok P",("Name","Surname"):"Pandey"},
    ("first_name","last_name"):{("Name","FullName"):"Rakesh P",("Name","Surname"):"Pandey"},
    ("first_name","last name"):{("Nmae","FullName"):"Suresh K",("Name","Surname"):"Pandey"}
}
dataframe=pd.DataFrame(tuple)
dataframe
```

```python
from collections import namedtuple
p1=namedtuple("Point","X Y")
data=pd.DataFrame([p1(0,1),p1(0,3)])
data
```

```python
p2=namedtuple("Point3D","X Y Z")
data=[p2(0,1,2),p2(3,4,5),p2(6,7,8)]
dataframe=pd.DataFrame(data)
dataframe
```

from_dict

```python
del dict
data=dict([("First",[1,2,3]),("Second",[4,5,6])])
dataframe=pd.DataFrame.from_dict(data)
dataframe
```

**LOGICAL OPERATOR**

```python
data={
    "a":[128,255,210],
    "b":[10,50,255],
}
data_2={
    "a":[128,255,210],
    "b":[10,50,255],
}
dataframe_1=pd.DataFrame(data)
dataframe_2=pd.DataFrame(data_2)

print("==:",dataframe_1==dataframe_2)
print("Equals:",dataframe_1.equals(dataframe_2))
print("Compare",dataframe_1.compare(dataframe_2))
result=(dataframe_1==dataframe_2).all().all()
print("Result:",result)
```

**TRANSPOSE**

```python
data={
    "Name":["a","b","c"],
    "Age":[25,24,22]
}
dataframe=pd.DataFrame(data)
dataframe
print(dataframe)
transposed_data=dataframe.T
transposed_data
```

```python
data={"values":[1,2,3,4]}
dataframe=pd.DataFrame(data)
print(dataframe)
transposed_data=dataframe.T
print(transposed_data)
```

```python
data={
    "ID":[1,2,3],
    "Name":["BOB","Marley","Alice"],
    "Age":[24,23,5]
}
dataframe=pd.DataFrame(data)
dataframe.set_index(["Name"],inplace=True)
dataframe.set_index(["ID"],drop=False)
dataframe.reset_index()
print(dataframe)
```
