```python
data={
    "Flight Number":["A101","C101","D101","E101","FC101"],
    "FROM":["DELHI","MUMBAI","CHENNAI","PUNE","PUNJAB"],
    "TO":["MUMBAI","DELHI","MUMBAI","PUNJAB","CHENNAI"],
    "D_Time":["08.00pm","4.30pm","5.40pm","6.40pm","9.00am"],
    "A_Time":["12:00pm","6:30pm","7.40pm","8.00pm","1.00pm"],
    "Airline":["Indigo","Vistara","Delta airlines","Kingfisher","United airlines"]
}
dates=pd.date_range("2025-01-27",periods=5,freq='D')
print(dates)
df3=pd.DataFrame(data,index=dates)
df3
```
